```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage

df = pd.read_csv(r"C:\Users\acer\Desktop\excelr\assignments\7 clustering\crime_data.csv")
X = df.drop('Unnamed: 0', axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform hierarchical clustering
agglomerative = AgglomerativeClustering(n_clusters=10, distance_threshold=None, linkage='ward')
agglomerative_labels = agglomerative.fit(X_scaled)

linkage_matrix = linkage(X_scaled, 'ward')


plt.figure(figsize=(12, 6))
dendrogram(linkage_matrix, truncate_mode='level', p=5)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')

print("Cluster Labels:", agglomerative_labels)
```

```python
from sklearn.metrics import silhouette_score

for n_clusters in range (1,49):
    agglomerative = AgglomerativeClustering(n_clusters=5)
    cluster_labels = agglomerative.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"Number of clusters: {n_clusters}, Silhouette Score: {silhouette_avg}")
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler 
from sklearn.cluster import AgglomerativeClustering, KMeans,DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score
```

```python
df1= df1.drop('East-West Airlines is trying to learn more about its customers.  Key issues are their',axis = 1)
```

```python
df1.columns
```

```python
 df1 = pd.get_dummies(df1, columns=['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'])
df1
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df1)

linkage_matrix = linkage(X_scaled, method='ward')  
dendrogram(linkage_matrix)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()

# K-Means Clustering
sse = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

plt.plot(range(2, 11), sse, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Method for Optimal k')
plt.show()

dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

# Evaluate clustering using Silhouette Score for K-Means
optimal_k = np.argmin(sse) + 2
kmeans_optimal = KMeans(n_clusters=optimal_k, random_state=42)
kmeans_labels = kmeans_optimal.fit_predict(X_scaled)
silhouette_kmeans = silhouette_score(X_scaled, kmeans_labels)

# Print results
print("Optimal Number of Clusters (Elbow Method):", optimal_k)
print("Silhouette Score for K-Means Clustering:", silhouette_kmeans)

# Visualize the clustering results 
df1['Hierarchical_Cluster'] = AgglomerativeClustering(n_clusters=optimal_k).fit_predict(X_scaled)
df1['KMeans_Cluster'] = kmeans_labels
df1['DBSCAN_Cluster'] = dbscan_labels

# Draw inferences from the clusters obtained
# Analyze and interpret the characteristics of each cluster based on the original features
# For example, calculate mean values of features for each cluster and analyze the differences
cluster_means = df1.groupby('KMeans_Cluster').mean()
print(cluster_means)
```
